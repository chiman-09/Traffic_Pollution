```python
import requests
url="https://www.aqi.in/dashboard/india/karnataka/bangalore"
page = requests.get(url)
print(page.status_code)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(page.content,"html.parser")
print(soup)
```

```python
all_locations = soup.find_all("a")
for rest in all_locations: 
    print(rest.text)
```
